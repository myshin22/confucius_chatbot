```python
INPUT_FILE_NAME = "analects_of_confucius.pickle"
OUTPUT_FILE_NAME = "analects_raw.txt"
```

```python
infile = open(INPUT_FILE_NAME, "r", encoding='UTF8')
outfile = open(OUTPUT_FILE_NAME, "w+", encoding='UTF8')

lines = infile.readlines()
for line in lines:
    if (line[0] == '<'):
        continue

    outfile.write(line)

infile.close()
outfile.close()
```
